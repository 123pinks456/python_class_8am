```python
info=""
file = open("info.csv","w")
n= int(input("enter n= "))
for i in range(n):
    name= input("enter name= ")
    age= int(input("enter age= "))
    add= input("enter add= ")
    info= info + f"{name},{age},{add}\n"

file.write(info)
file.close()
```
